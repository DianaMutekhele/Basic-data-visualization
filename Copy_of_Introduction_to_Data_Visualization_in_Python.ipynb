## Introduction
There are two main branches of visualization that we can achieve using Python libraries;
1. Static Visualization - Examples include; Matplotlib, Pandas, Seaborn

1. Interactive Visualization - eg. Plotly, Bokeh etc

Below is an overview of the visualization libraries;


*   [Matplotlib](https://matplotlib.org/): low level, provides lots of freedom
*   [Pandas Visualization](https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html): easy to use interface, built on Matplotlib and has an in-built support for pandas dataframes and series
*   [Seaborn](https://seaborn.pydata.org/): high-level interface, great default styles. Also built on Matplotlib.
*   [Plotly](https://plotly.com/python/): Useful for interactive graphing

## Matplotlib
### Scatter Plot
To create scatter plot we will use the Matplotlib method '*scatter*'. We will also use *plt.subplots* features to give the plot titles and lables

```python
# create a figure and axis
fig, ax = plt.subplots()

# scatter the sepal_length against the sepal_width
ax.scatter(iris['sepal_length'], iris['sepal_width'])
# set a title and labels
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

```python
iris['class'].unique()
```

```python
#To create a color dictionary

colors = {'Iris-setosa':'r', 'Iris-versicolor':'g', 'Iris-virginica':'b'}
# create a figure and axis
fig, ax = plt.subplots()
# plot each data-point
for i in range(len(iris['sepal_length'])):
    ax.scatter(iris['sepal_length'][i], iris['sepal_width'][i],color=colors[iris['class'][i]])
# set a title and labels
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

### Line Chart
We will use the plot method to *plot* line graphs.

```python
# get columns to plot
columns = iris.columns.drop(['Id','class'])
# create x data
x_data = range(0, iris.shape[0])
# create figure and axis
fig, ax = plt.subplots()
# plot each column
for column in columns:
    ax.plot(x_data, iris[column], label=column)
# set title and legend
ax.set_title('Iris Dataset')
ax.legend()
```

### Histogram
We will use *hist* method. Matplotlib automatically calculate how often each class occurs incase of categorical data.

```python
# create figure and axis
fig, ax = plt.subplots()
# plot histogram
ax.hist(wine_reviews['points'])
# set title and labels
ax.set_title('Wine Review Scores')
ax.set_xlabel('Points')
ax.set_ylabel('Frequency')
```

### Bar Chart
A bar chart can be created using the bar method. The bar-chart isn’t automatically calculating the frequency of a category so we are going to use pandas value_counts function to do this.

```python
# create a figure and axis 
fig, ax = plt.subplots() 
# count the occurrence of each class 
data = wine_reviews['points'].value_counts() 
# get x and y data 
points = data.index 
frequency = data.values 
# create bar chart 
ax.bar(points, frequency) 
# set title and labels 
ax.set_title('Wine Review Scores') 
ax.set_xlabel('Points') 
ax.set_ylabel('Frequency')
```

### Heatmap


```python
# get correlation matrix

corr = iris.drop(['Id'], axis=1).corr()
fig, ax = plt.subplots()
# create heatmap
im = ax.imshow(corr.values)

# set labels
ax.set_xticks(np.arange(len(corr.columns)))
ax.set_yticks(np.arange(len(corr.columns)))
ax.set_xticklabels(corr.columns)
ax.set_yticklabels(corr.columns)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(corr.columns)):
    for j in range(len(corr.columns)):
        text = ax.text(j, i, np.around(corr.iloc[i, j], decimals=2),
                       ha="center", va="center", color="black")
```

## Pandas Visualization
Pandas Visualization makes it really easy to create plots out of a pandas dataframe and series. It also has a higher level API than Matplotlib and therefore we need less code for the same results.

###Scatter Plot

```python
iris.plot.scatter(x='sepal_length', y='sepal_width', title='Iris Dataset')
```

### Line Chart

```python
iris.drop(['Id','class'], axis=1).plot.line(title='Iris Dataset')
```

### Histogram
We will use the *plot.hist* method to plot a histogram

```python
#Histogram
wine_reviews['points'].plot.hist()
```

```python
iris.columns
```

To create multiple histograms, we will set subplot argument to be True.


```python
iris.drop(['Id','class'], axis=1).plot.hist(subplots=True, layout=(2,2), figsize=(10, 10), bins=20)
```

### Bar Chart
To plot a bar-chart we can use the *plot.bar()* method, but before we can call this we need to get our data. For this we will first count the occurrences using the *value_count()* method and then sort the occurrences from smallest to largest using the *sort_index()* method.

```python
wine_reviews['points'].value_counts().sort_index().plot.bar()
```

We can use *plot.barh()* method to convert the above graph into an horizontal bar-chart

```python
wine_reviews['points'].value_counts().sort_index().plot.barh()
```

We can also plot other data then the number of occurrences.

```python
wine_reviews.groupby("country").price.mean().sort_values(ascending=False)[:5].plot.bar()
```

## Seaborn
### Scatter plot
We will use *.scatter()* method to plot a scatter plot

```python
sns.scatterplot(x='sepal_length', y='sepal_width', data=iris)
```

To highlight the points by class we will use *hue* argument

```python
sns.scatterplot(x='sepal_length', y='sepal_width', hue='class', data=iris)
```

### Line chart
We will use the *lineplot()* method. We can also use *kdeplot()* method to achieve the same effect

```python
sns.lineplot(data=iris.drop(['Id','class'], axis=1))
```

### Histogram
We use *displot()* method in seaborn library to create histograms as below

```python
sns.displot(wine_reviews['points'], bins=20, kde=False)
```

```python
sns.displot(wine_reviews['points'], bins=10, kde=True)
```

### Bar chart
In Seaborn a bar-chart can be created using the sns.countplot method and passing it the data.

```python
sns.countplot(x=wine_reviews['points'])
```

### Box plots
We can create box plots using seaborns sns.boxplot method and passing it the data as well as the x and y column name.

```python
df = wine_reviews[(wine_reviews['points']>=95) & (wine_reviews['price']<1000)]
sns.boxplot('points', 'price', data=df)
```

### Heatmap (Correlation matrix)


```python
sns.heatmap(iris.drop(['Id'], axis=1).corr(), annot=True)
```
